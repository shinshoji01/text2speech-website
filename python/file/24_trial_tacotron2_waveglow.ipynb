```python
import torch
import numpy as np
from scipy.io.wavfile import write
import matplotlib.pyplot as plt
from WaveGlow import call_waveglow
import IPython
```

```python
def play_audio(data, rate):
    IPython.display.display(IPython.display.Audio(data=data,rate=rate))
```

# Prepare

```python
# ----- adjustable parameters ----- #
save_path = "./parameters/waveglow_official.pth"
# --------------------------------- #

waveglow = torch.hub.load('nvidia/DeepLearningExamples:torchhub', 'nvidia_waveglow', pretrained=True, map_location="cuda")
waveglow = waveglow.remove_weightnorm(waveglow)
waveglow = waveglow.to("cuda")

torch.save(waveglow.state_dict(), save_path)
```

```python
model_path = "./parameters/waveglow_official.pth"
waveglow = call_waveglow(model_path)
waveglow = waveglow.to('cuda')
waveglow.eval()
print()
```

```python
tacotron2 = torch.hub.load('nvidia/DeepLearningExamples:torchhub', 'nvidia_tacotron2')
tacotron2 = tacotron2.to('cuda')
tacotron2.eval()
print()
```

# Speech Synthesis

```python
text = "My name is Sho Inoue"
```

```python
utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tts_utils')
sequences, lengths = utils.prepare_input_sequence([text])
```

```python
with torch.no_grad():
    mel, _, _ = tacotron2.infer(sequences, lengths)
    audio = waveglow.infer(mel)
audio_numpy = audio[0].data.cpu().numpy()
rate = 22050
```

```python
play_audio(audio_numpy[0], rate)
```

```python
text = "My name is Sho Inoue"
rate = 22050
```

```python
import time
```

```python
sample_num = 10
time_list = []
for _ in range(sample_num):
    start = time.time()
    # preparation
    model_path = "./parameters/waveglow_official.pth"
    waveglow = call_waveglow(model_path)
    waveglow = waveglow.to('cuda')
    waveglow.eval()

    tacotron2 = torch.hub.load('nvidia/DeepLearningExamples:torchhub', 'nvidia_tacotron2')
    tacotron2 = tacotron2.to('cuda')
    tacotron2.eval()

    # speech generation
    utils = torch.hub.load('NVIDIA/DeepLearningExamples:torchhub', 'nvidia_tts_utils')
    sequences, lengths = utils.prepare_input_sequence([text])

    with torch.no_grad():
        mel, _, _ = tacotron2.infer(sequences, lengths)
        audio = waveglow.infer(mel)
    audio_numpy = audio[0].data.cpu().numpy()
    time_list.append(time.time()-start)
np.mean(time_list)
```

```python
play_audio(audio_numpy[0], rate)
```
